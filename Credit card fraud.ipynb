```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import os
```

```python
#loading the dataset
filepath= os.path.join(os.getcwd(),'creditcard.csv')
dataset=pd.read_csv(filepath)
```

```python
#Checking the first 5 rows and knowing better about the data
dataset.head()
```

```python
#Checking the tail of the dataset
dataset.tail()
```

```python
#Describing the dataset
dataset.describe()
```

```python
#More Info 
dataset.info()
```

```python
# distribution of legit transactions & fraudulent transactions
dataset['Class'].value_counts()
# 0 ---> Normal
# 1 ---> fraud
```

```python
# separating the data for analysis
legit = dataset[dataset.Class == 0]
fraud = dataset[dataset.Class == 1]
```

```python
#Checking for the shape of each, knowing that the data is unbalanced
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
dataset.groupby('Class').mean()
```

```python
#we will reduce the number of legit trasncation so the model don't overfit by using a technique called "Under Sampling"
legit_sample=legit.sample(n=492)
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
#Counting the new dataset
new_dataset['Class'].value_counts()new_dataset.groupby('Class').mean()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
y = new_dataset['Class']
```

```python
#Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### Trainning the model

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
